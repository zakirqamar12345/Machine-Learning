```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
plt.xlabel('area (sqr ft)', fontsize = 12)
plt.ylabel('price(US $)', fontsize = 12)
plt.scatter(df.area, df.price, color = 'red', marker = '*')
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['area']], df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3300]])
```

```python
plt.xlabel('area (sqr ft)', fontsize = 12)
plt.ylabel('price(US $)', fontsize = 12)
plt.scatter(df.area, df.price, color = 'red', marker = '*')
plt.plot(df.area, reg.predict(df[['area']]), color = 'blue', marker = '+')
```

```python
df_new = pd.read_csv('areas.csv')
df_new
```

```python
d= reg.predict(df_new)
d
```

```python
df_new['price'] = d
```
